```python
import pandas
csv_data = pandas.read_csv('C:/Users/yje20/OneDrive/문서/데분기/프로젝트 준비/road.csv',encoding='cp949')

csv_spring= csv_data[(csv_data.발생월>2) & (csv_data.발생월<6)]
csv_sprAcc = csv_spring[['사고건수']]
csv_summer= csv_data[(csv_data.발생월>5) & (csv_data.발생월<9)]
csv_sumAcc = csv_summer[['사고건수']]
csv_fall= csv_data[(csv_data.발생월>8) & (csv_data.발생월<12)]
csv_fallAcc = csv_fall[['사고건수']]
csv_winter= csv_data[(csv_data.발생월>11) | (csv_data.발생월<3)]
csv_winAcc = csv_winter[['사고건수']]
```

```python
import numpy as np
sumArr = pandas.Series([np.sum(csv_sprAcc,0),np.sum(csv_sumAcc,0),
                        np.sum(csv_fallAcc,0),np.sum(csv_winAcc,0)])
sea = pandas.Series([1,2,3,4])
seasonAcc = pandas.DataFrame(data = dict(season = sea, Accidents = sumArr))
```

```python
seasonAcc['Accidents'].astype(float).head(4).plot.bar() #계절별 사고 건수 분류
```

```python
spr_cp = csv_spring[csv_spring.사고유형대분류 == '차대사람']
spr_cc = csv_spring[csv_spring.사고유형대분류 == '차대차']
spr_co = csv_spring[csv_spring.사고유형대분류 == '차량단독']

sum_cp = csv_summer[csv_summer.사고유형대분류 == '차대사람']
sum_cc = csv_summer[csv_summer.사고유형대분류 == '차대차']
sum_co = csv_summer[csv_summer.사고유형대분류 == '차량단독']

fall_cp = csv_fall[csv_fall.사고유형대분류 == '차대사람']
fall_cc = csv_fall[csv_fall.사고유형대분류 == '차대차']
fall_co = csv_fall[csv_fall.사고유형대분류 == '차량단독']

win_cp = csv_winter[csv_winter.사고유형대분류 == '차대사람']
win_cc = csv_winter[csv_winter.사고유형대분류 == '차대차']
win_co = csv_winter[csv_winter.사고유형대분류 == '차량단독']
```

```python
spr_cp_num = spr_cp[['사고건수']]
spr_cc_num = spr_cc[['사고건수']]
spr_co_num = spr_co[['사고건수']]

sum_cp_num = sum_cp[['사고건수']]
sum_cc_num = sum_cc[['사고건수']]
sum_co_num = sum_co[['사고건수']]

fall_cp_num = fall_cp[['사고건수']]
fall_cc_num = fall_cc[['사고건수']]
fall_co_num = fall_co[['사고건수']]

win_cp_num = win_cp[['사고건수']]
win_cc_num = win_cc[['사고건수']]
win_co_num = win_co[['사고건수']]
```

```python
cnp = pandas.Series([np.sum(spr_cp_num,0),np.sum(sum_cp_num,0),np.sum(fall_cp_num,0),np.sum(win_cp_num,0)])
cnc = pandas.Series([np.sum(spr_cc_num,0),np.sum(sum_cc_num,0),np.sum(fall_cc_num,0),np.sum(win_cc_num,0)])
conly = pandas.Series([np.sum(spr_co_num,0),np.sum(sum_co_num,0),np.sum(fall_co_num,0),np.sum(win_co_num,0)])
season = pandas.Series([1,2,3,4])

seasonAccKind = pandas.DataFrame( data = dict(carNpeo = cnp, carNcar = cnc
                                             , caronly = conly, seasons = season))

seasonAccKind[['carNpeo']].astype(float).head(4).plot.bar(color = 'red')
#계절별 차대사람 사고건수

seasonAccKind[['carNcar']].astype(float).head(4).plot.bar(color = 'orange')
#계절별 차대차 사고건수

seasonAccKind[['caronly']].astype(float).head(4).plot.bar(color = 'green')
#계절별 차량단독 사고건수
```
